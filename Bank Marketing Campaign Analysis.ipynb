```python
def age(dataframe):
    dataframe.loc[dataframe['age'] <= 32, 'age'] = 1
    dataframe.loc[(dataframe['age'] > 32) & (dataframe['age'] <= 47), 'age'] = 2
    dataframe.loc[(dataframe['age'] > 47) & (dataframe['age'] <= 70), 'age'] = 3
    dataframe.loc[(dataframe['age'] > 70) & (dataframe['age'] <= 98), 'age'] = 4
           
    return dataframe

age(bank_data);
```


<b> 
<font color="red">Upsampling of minority</font>
</b>

```python
from sklearn.utils import resample
```

```python
X = pd.concat([X_train, y_train], axis=1) 
```

```python
target_yes = X[X['target']==1]
target_no = X[X['target']==0]
```

```python
target_yes_upsampled = resample(target_yes,replace=True, n_samples=len(target_no), random_state=27)
```

```python
upsampled = pd.concat([target_yes_upsampled, target_no])
```

```python
upsampled[upsampled['target']==1].count()
```

```python
X_train_up = upsampled.drop('target',axis=1)
y_train_up = upsampled['target']
```

<b> 
<font color="red">Logistic Regression on Upsampled data</font>
</b>

```python
log_up = LogisticRegression()
log_up.fit(X_train_up, y_train_up)
```

```python
pred_up = log_up.predict(X_test)
```

```python
print(classification_report(y_test, pred_up))
print(confusion_matrix(y_test, pred_up))
```

<b> 
<font color="red">Random Forest on Upsampled data</font>
</b>

```python
rfc_up = RandomForestClassifier(n_estimators=5000)
```

```python
rfc_up.fit(X_train_up, y_train_up)
pred_rfc_up = rfc_up.predict(X_test)
```

```python
print(classification_report(y_test, pred_rfc_up))
print(confusion_matrix(y_test, pred_rfc_up))
```

<b> 
<font color="green">Downsampling of majority</font>
</b>

```python
target_no_downsampled = resample(target_no,replace=True, n_samples=len(target_yes), random_state=27)
```

```python
downsampled = pd.concat([target_yes, target_no_downsampled])
```

```python
X_train_down = downsampled.drop('target', axis=1)
```

```python
y_train_down = downsampled['target']
```
